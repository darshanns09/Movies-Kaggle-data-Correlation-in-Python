```python
numeric_df = df.select_dtypes(include='number')

numeric_df.corr(method ='pearson')   # pearson,spearman,kendall
```

```python
# Correlation Matrix between all numeric columns


numeric_df.corr(method ='spearman')
```

```python
correlation_matrix = numeric_df.corr(method ='pearson') 

sns.heatmap(correlation_matrix , annot =True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Using factorize - this assigns a random numeric value for each unique categorical value

correlation_mat = df.apply(lambda x: x.factorize()[0]).corr(method ='pearson') 
correlation_mat
```

```python
sns.heatmap(correlation_mat , annot =True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
corr_pairs  = correlation_mat.unstack().sort_values()
```

```python
high_corr = corr_pairs[(corr_pairs) > 0.5]
high_corr
```
